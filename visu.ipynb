```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("vendas_mensais.csv")
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["Mês"], df["Vendas"], marker="o", linestyle="-", color="b", label="Vendas")

plt.xlabel("Mês")
plt.ylabel("Vendas (R$)")
plt.title("Vendas Mensais")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.show()
```
